```python
import os
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
from pprint import pprint
import csv
import datetime
import numpy as np
```

## Pulled in csv from main youtube data script

```python
read_csv = pd.read_csv("Music_Videos.csv")
df = pd.DataFrame(read_csv)
df["Release Date"] = df["Release Date"].apply(lambda x: f"{x}-2020")
df.head()
```

### Converting "Release Date" Column from string object to datetime object

```python
df["Release Date"] = df["Release Date"].apply(lambda x: datetime.datetime.strptime(x,"%m-%d-%Y"))
df["Release Date"] = df["Release Date"].apply(lambda x: datetime.datetime.strftime(x, '%m-%d (%a)'))
df.head()
```

### The following table shows:
### The top rated music video for each release date, sorted by viewcount in descending order.

```python
release_date_max = df.groupby(df["Release Date"]).max().sort_values(by="View count", 
                                                        ascending = False)
release_date_max = release_date_max.drop(columns="Unnamed: 0")
top_four_list = ["07-23 (Thu)", "07-24 (Fri)", "07-17 (Fri)", "07-30 (Thu)"]
release_date_max.style.apply(lambda x: ['background: yellow' if x.name in top_four_list 
                                        else '' for i in x], axis=1)
```

### Here we see that the top 4 most viewed videos out of the 50 highest rated were all released on either Thursday or Friday.
